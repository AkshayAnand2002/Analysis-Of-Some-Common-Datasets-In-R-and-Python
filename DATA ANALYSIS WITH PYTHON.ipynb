```python
##Use subplots to create a figure with two separate line plots showing the number of 
##marriages and divorces per capita in the U.S. between 1867 and 2014. Place the two 
##line plots on top of each other so the years line up.
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)## it has 2 rows, 1 column and it is first element of the 2 elements.
plt.plot(data['Year'], data['Marriages_per_1000'])
plt.xlabel('Year')
plt.ylabel('Per capita count')
plt.title('Marriages between 1867 and 2014')
plt.subplot(2,1,2) # 2nd element of the 2 elements.
plt.plot(data['Year'], data['Divorces_per_1000'])
plt.xlabel('Year')
plt.ylabel('Per capita count')
plt.title('Divorces between 1867 and 2014')
plt.show()
```

```python
##Create two box plots to compare the earnings of recent college graduates in majors 
##that are primarily composed of women to the earnings of recent college graduates in 
##majors that are primarily composed of men
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('C:/college-grads-earnings.csv')
men = data[data['ShareWomen']<0.5] ## since women's share <0.5 so it is composed primarily of men.
women = data[data['ShareWomen']>0.5]## women's share > 0.5 so it is primarily composed of women.
plot_data = [men['Median'],women['Median']]## calculated median to get best value for data representation.
## Also we require median for drawing boxplot.
plt.boxplot(plot_data,notch=True)
plt.title('Comparison of (median) salaries of majors primarily composed of men and women')
plt.show()
```

```python
## Show barplots for Marriages and Divorces per 1000 from  us marriage divorce dataset from year 1991 to 2000.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data= pd.read_csv('C:/us-marriages-divorces-1867-2014.csv')
data.dropna(inplace=True)###for permanent removal of null values from data
data=data[(data.Year>=1991)&(data.Year<=2000)]
years=data['Year'].values
marriage=data['Marriages_per_1000'].values
divorce=data['Divorces_per_1000'].values
plt.bar(years,marriage,0.4,label="Marriage",color="g")
plt.bar(years,divorce,0.4,label="Divorce",color='r') 
plt.xlabel("Years")
plt.ylabel("No. Of Couples")
plt.title("Marriages and Divorces per 1000 in the US")
plt.legend()
plt.show()
## if we don't space the 2 barplots we get stacked barplots.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data= pd.read_csv('C:/us-marriages-divorces-1867-2014.csv')
data.dropna(inplace=True)
data=data[(data.Year>=1991)&(data.Year<=2000)]
years=data['Year'].values
marriage=data['Marriages_per_1000'].values
divorce=data['Divorces_per_1000'].values
plt.bar(years-0.2,marriage,0.4,label="Marriage",color="g")## years-0.2 to left space barplot to avoid overlapping
plt.bar(years+0.2,divorce,0.4,label="Divorce",color='r') ## years+0.2 to rightspace to again avoid overlapping
# 0.4 is width of bars we can increase or decrease them by increasing or decreasing the value.
plt.xlabel("Years")
plt.ylabel("Couples")
plt.title("Marriages and Divorces per 1000 in the US")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data= pd.read_csv('C:/actor_kill_counts.csv')
data.dropna(inplace=True)
data=data.sort_values('Count',ascending=True)
actor= data['Actor'].values
kills= data['Count'].values
left=actor
height=kills
plt.barh(left,height) ## Horizontal barplot with height as kill count and left as actors i.e. base of bar has actor's names
plt.show()
```
