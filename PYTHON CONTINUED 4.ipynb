```python
##joint plots of profit and sales using jointflot function
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
market_fact=pd.read_csv("C:/RFILES/market_fact.csv")
sns.jointplot('Sales','Profit',market_fact)
plt.show()
```

```python
sns.jointplot(market_fact['Sales'],market_fact['Profit'])
plt.show()
```

```python
market_fact1=market_fact[(market_fact.Profit < 10000) & (market_fact.Sales < 20000)]
sns.jointplot('Sales','Profit',market_fact1)
plt.savefig('scatter')
plt.show()
```

```python
#using kind=hex and colors and data is skewed here.
sns.jointplot('Sales','Profit',market_fact1,kind="hex",color="y")
plt.show()#####output is like hexagon.
```

```python
market_fact2=market_fact[(market_fact.Profit < 100) & (market_fact.Sales < 200)]
sns.jointplot('Sales','Profit',market_fact2,kind="hex",color='c')
plt.show()
```

```python
##instead of plotting columnwise we can also plot the relationship among all the columns.
####this is called cor-relation matrix.
sns.pairplot(market_fact2)
plt.show()
plt.savefig('corrmatrix')
```
