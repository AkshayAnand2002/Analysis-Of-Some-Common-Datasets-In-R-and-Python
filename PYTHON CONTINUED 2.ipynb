```python
FIFA_2018=pd.DataFrame({'FIFA_TEAM':['BEL','FRANCE','CROTIA'],'MATCHES PLAYED':[7,7,7],'GOALS SCORED':[11,10,9]})
FIFA_2018.set_index('FIFA_TEAM',inplace=True)
FIFA_2018
```

```python
FIFA_2014=pd.DataFrame({'FIFA_TEAM':['BEL','FRANCE','CROTIA','KOREA'],'MATCHES PLAYED':[7,6,5,4],'GOALS SCORED':[15,40,19,9]})
FIFA_2014.set_index('FIFA_TEAM',inplace=True)
FIFA_2014
```

```python
TOTAL=FIFA_2014+FIFA_2018
TOTAL
```

```python
# THE fill_value arguement in the df.add() replaces the NaN value with 0.
TOTAL=FIFA_2018.add(FIFA_2014,fill_value=0)
TOTAL
```

```python
mar_cust=pd.merge(market_fact,cust_dimen,how='inner',on='Cust_id')
mar_cust_pr=pd.merge(mar_cust,prod_dimen,how='inner',on='Prod_id')
mar_cust_pr_ship=pd.merge(mar_cust_pr,shipping_dimen,how='inner',on='Ship_id')
mar_cust_pr_ship_ord=pd.merge(mar_cust_pr_ship,orders_dimen,how='inner',on='Ord_id')
mar_cust_pr_ship_ord.head()
```

```python
def sales_500(x):
    if x>500:
        return True
    elif x<=500:
        return False 
#apply function returns value'Y' if sales>500 else 'N'.
```

```python
mar_cust_pr_ship_ord['sales_500']=mar_cust_pr_ship_ord['Sales'].apply(sales_500)
mar_cust_pr_ship_ord.head()## sew last column of the output which is sales_500 which gives True or False .
```

```python
mar_cust_pr_ship_ord['sales_500']=mar_cust_pr_ship_ord['Sales'].apply(lambda x: True if x > 500 else False)
mar_cust_pr_ship_ord.head()##above program done by lambda function.
```

```python
#use lambda function to modify existing data.
#using round function to remove decimals. in the Sales column
mar_cust_pr_ship_ord['Sales']=mar_cust_pr_ship_ord['Sales'].apply(lambda x: round(x))
mar_cust_pr_ship_ord.head()
```
