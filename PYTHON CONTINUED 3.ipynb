```python
import numpy as np
import pandas as pd
estate=pd.read_csv('C:/RFILES/realestatedata.csv')
estate.head()
```

```python
estate.isnull()#to check null values are there or not in different rows and columns by True or False.
```

```python
estate.isnull().sum()#total values which are null columnwise.
```

```python
estate.isnull().any()#columns having atleast1 null value.
```

```python
#getting missing value columnwise .
estate.isnull().any(axis=0)#column-wise
```

```python
#getting missing value row-wise.
estate.isnull().any(axis=1)#row-wise
```

```python
#getting total no. of columns with missing values.
estate.isnull().any(axis=0).sum()
```

```python
#getting total no. of rows with missing values.
estate.isnull().any(axis=1).sum()
```

```python
#getting rows with all values null
estate.isnull().all(axis=1).sum()
```

```python
#getting columnwise % of null.
round(100*(estate.isnull().sum()/len(estate.index)),2)
```

```python
# as missing value% >30 so droop those columns.
estate=estate.drop('BuildingArea',axis=1)
estate=estate.drop('YearBuilt',axis=1)
estate=estate.drop('CouncilArea',axis=1)
```

```python
round(100*(estate.isnull().sum()/len(estate.index)),2)
```

```python
estate[estate.isnull().sum(axis=1) >5] #analysing rowwise and check if column values missing are more than 5.
```

```python
len(estate[estate.isnull().sum(axis=1) >5].index)# counting no. of rows with more than 5 column values missing.
```

```python
#how much % is this.
100*(len(estate[estate.isnull().sum(axis=1) >5].index)/len(estate.index))
```

```python
#retaining rows having <=5 NaNs
estmate=estate[estate.isnull().sum(axis=1) <=5]
#looking at the summary once again.
round(100*(estate.isnull().sum()/len(estate.index)),2)
```

```python
#price column has 21% missing values.As price is important to our study we will remove the rows with NaN price value in them.
#removing NaN Price Rows.
estate=estate[~np.isnan(estate['Price'])]
round(100*(estate.isnull().sum()/len(estate.index)),2) 
```

```python
estate['Landsize'].describe()#here considerable variation between min,max,25th percentile and 75th percentile.
#hence,safe to remove rows having NaN in Landsize. 
```

```python
#removing NaNs in Landside.
estate=estate[~np.isnan(estate['Landsize'])]
round(100*(estate.isnull().sum()/len(estate.index)),2) 
```

```python
estate[~np.isnan(estate['Lattitude'])]
```

```python
estate.loc[:,['Lattitude','Longtitude']].describe()
```

```python
#from above the difference between mean and median is very less. also difference between 25th and 75th percentile is very less.
#so its good indicator that if values are replaced then data will not be biased.
estate.loc[np.isnan(estate['Lattitude']),['Lattitude']]=estate['Lattitude'].mean()
round(100*(estate.isnull().sum()/len(estate.index)),2)
```

```python
estate.loc[np.isnan(estate['Longtitude']),['Longtitude']]=estate['Longtitude'].mean()
round(100*(estate.isnull().sum()/len(estate.index)),2)
```

```python
estate.loc[:,['Bathroom','Car']].describe()
```

```python
#we can't replace using mean,median,25th and 75th percentile even though variation is less because car mean is 1.6  .
#finding frequency of each value and replacing , note that we are converting columns into category.
estate['Car']=estate['Car'].astype('category')
estate['Car'].value_counts()
```

```python
#most common valueis 2 hence replacing the value.
estate.loc[pd.isnull(estate['Car']),['Car']]=2
```

```python
estate['Bathroom']=estate['Bathroom'].astype('category')
estate['Bathroom'].value_counts()
```

```python
estate.loc[pd.isnull(estate['Bathroom']),['Bathroom']]=1
```

```python
round(100*(estate.isnull().sum()/len(estate.index)),2)
#cleaning of data completed.
```
